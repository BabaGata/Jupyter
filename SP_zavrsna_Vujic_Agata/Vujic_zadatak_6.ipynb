Konstruirajte funkciju $\texttt{walk1D()}$ tako da simulira tzv. jednodimenzionalnog nasumičnog šetača (random walker). Svaki šetačev korak treba biti iste, jedinične
duljine, ulijevo ili udesno. Dakle, u svakom koraku trebamo slučajan odabir
koraka +1 ili -1. Šetač kreće iz ishodišta i napravi ${N = 10^3}$ koraka. Zabilježite
konačni položaj šetača. Ponovite proceduru ${10^4}$ puta. Nacrtajte histogram
konačnih položaja šetača. Uvjerite se u ispravnost tvrdnje da je prosječna udaljenost nasumičnog jednodimenzionalnog šetača od ishodišta nakon N koraka
nula, prosječna kvadratna udaljenost N, sa standardnom devijacijom ${\sqrt{2}N}$, dok
je prosječna apsolutna udaljenost
${\sqrt{2N/\pi}}$.
Na posebnom grafu nacrtajte jednu šetačevu šetnju, odnosno udaljenost od
početnog položaja šetača u ovisnosti o broju koraka.

```python
import numpy as np
from scipy import *
import matplotlib.pyplot as plt


def walk1D():
    x = 0
    put = []
    for i in range(10**3):
        r = np.random.choice([-1, 1])
        x = x + r
        put.append((x, i))
    return put

konacnipolozaj = np.array([np.array(walk1D())[len(walk1D())-1, 0] for j in range(10**4)])
kvadratpolozaja = konacnipolozaj**2
print('Prosjecna udaljenost je:', np.average(konacnipolozaj))
print('Prosjecna kvadratna udaljenost:', np.average(kvadratpolozaja), 'N =', len(walk1D()))
print('Standardna devijacija kvadratne udaljenosti:', np.std(kvadratpolozaja), 'Std prema formuli:', np.sqrt(2)*len(walk1D()))
print('Prosjecna apsolutna udaljenost:', np.average(np.absolute(konacnipolozaj)),
      'Prema formuli:', np.sqrt(2*len(walk1D())/pi))
fig, ax = plt.subplots()
ax.hist(konacnipolozaj)

fig2, ax2 = plt.subplots()
ax2.plot(np.array(walk1D())[:, 1], np.array(walk1D())[:,0])
ax2.set_xlabel('N')
ax2.set_ylabel('Udaljenost od sredista')
```
