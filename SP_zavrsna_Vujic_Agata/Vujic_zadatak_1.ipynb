### Zadatak 1
Zadana je funkcija
$${f(x, y) = \frac{5x^2y^2}{x^2 + y^2}}$$
a) Defnirajte funkciju i izračunajte vrijednost funkcije u točki (1, 1). Je li
funkcija defnirana u točki (0, 0)?
<br>b) Nacrtajte plohu ${z = f(x, y)}$ na pravokutniku ${[-1, 1] \times [-1, 1]}$.

```python
import matplotlib.pyplot as plt
import numpy as np

def funkcija (x, y):
    return (5 * x**2 * y**2)/(x**2 + y**2)
print('Funkcija u točki (1, 1) je', funkcija(1, 1))
print('Funkcija nije definirana u točki (0, 0)')

t = np.linspace(-1, 1)
X, Y = np.meshgrid(t, t)
Z = funkcija(X, Y)
fig = plt.figure(figsize=[10, 10])
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis', edgecolor='none')
ax.set_xlabel('x os', fontsize=15)
ax.set_ylabel('y os', fontsize=15)
ax.set_zlabel('f(x,y)', fontsize=15)
```
