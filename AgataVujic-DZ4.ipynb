### Zadatak 1
Uočite da se funkcija $\texttt{fibBinet(x)}$ iz <font color=blue>prošlog odjeljka</font> može izvrijedniti i za vrijednosti argumenta
𝑥 koje nisu cjelobrojne. Nacrtajte graf te funkcije za −4 < 𝑥 < 7 i superponirajte na njega
(u drugoj boji) točke koje odgovaraju vrijednostima funkcije za pozitivne cjelobrojne argumente
[(1, F_1), (2, F_2), ... (7, F_7)].

```python
from scipy.constants import golden_ratio, pi
import numpy as np
def fibBinet(n):
    a = golden_ratio**n - np.cos(n*pi)*(golden_ratio**(-n))
    d = np.sqrt(5)
    F = a/d
    return F

x = np.linspace(-4, 7)
xn = np.arange(1,8,1)

import matplotlib.pyplot as plt

plt.plot(x, fibBinet(x))
plt.scatter(xn, fibBinet(xn), color = 'red')
plt.xlabel('x', fontsize = 20)
plt.ylabel('fibBinet(x)', fontsize = 20)
```

### Zadatak 2
Koristeći trigonometrijske funkcije nacrtajte kružnicu kao parametarsku krivulju.

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 2*np.pi)
plt.plot(np.cos(t), np.sin(t))
plt.title('Kružnica', fontsize = 20)
plt.xlabel('x', fontsize = 20)
plt.ylabel('y', fontsize = 20)
```

### Zadatak 4
Nacrtajte kružnicu u kompleksnoj ravnini zadanu kompleksnim brojevima
$${ \{z = \rho𝑒^{i\eta}(1 + 𝑒^{i\phi}) − 1 \space\space\space\space|\space\space\space\space\phi ∈ [0, 2\pi),\space \rho = 1.3,\space \eta = 0.2\} }$$
te krivulju koja se dobije kad se ova kružnica podvrgne <font color=green>transformaciji Žukovskog</font>:
$${z\rightarrow\omega=\frac{1}{2}(z+\frac{1}{z})}$$

```python
import numpy as np
import matplotlib.pyplot as plt

def z (ro, fi, eta):
    a = ro * (np.e ** (1j * eta))
    b = 1 + (np.e ** (1j*fi))
    c = a * b - 1
    return c
def w(z):
    a = (1/2) * (z + (1/z))
    return a
fi = np.linspace(0, 2*np.pi)
ro = 1.3
eta = 0.2

plt.plot(z(ro, fi, eta).real, z(ro, fi, eta).imag)
plt.plot(w(z(ro, fi, eta)).real, w(z(ro, fi, eta)).imag)
```
