### Zadatak 1
Uoƒçite da se funkcija $\texttt{fibBinet(x)}$ iz <font color=blue>pro≈°log odjeljka</font> mo≈æe izvrijedniti i za vrijednosti argumenta
ùë• koje nisu cjelobrojne. Nacrtajte graf te funkcije za ‚àí4 < ùë• < 7 i superponirajte na njega
(u drugoj boji) toƒçke koje odgovaraju vrijednostima funkcije za pozitivne cjelobrojne argumente
[(1, F_1), (2, F_2), ... (7, F_7)].

```python
from scipy.constants import golden_ratio, pi
import numpy as np
def fibBinet(n):
    a = golden_ratio**n - np.cos(n*pi)*(golden_ratio**(-n))
    d = np.sqrt(5)
    F = a/d
    return F

x = np.linspace(-4, 7)
xn = np.arange(1,8,1)

import matplotlib.pyplot as plt

plt.plot(x, fibBinet(x))
plt.scatter(xn, fibBinet(xn), color = 'red')
plt.xlabel('x', fontsize = 20)
plt.ylabel('fibBinet(x)', fontsize = 20)
```

### Zadatak 2

### Zadatak 4
Nacrtajte kru≈ænicu u kompleksnoj ravnini zadanu kompleksnim brojevima
$${ \{z = \rhoùëí^{i\eta}(1 + ùëí^{i\phi}) ‚àí 1 \space\space\space\space|\space\space\space\space\phi ‚àà [0, 2\pi),\space \rho = 1.3,\space \eta = 0.2\} }$$
te krivulju koja se dobije kad se ova kru≈ænica podvrgne <font color=green>transformaciji ≈Ωukovskog</font>:
$${z\rightarrow\omega=\frac{1}{2}(z+\frac{1}{z})}$$

```python
import numpy as np
import matplotlib.pyplot as plt

def z (ro, fi, eta):
    a = ro * (np.e ** (1j * eta))
    b = 1 + (np.e ** (1j*fi))
    c = a * b - 1
    return c
def w(z):
    a = (1/2) * (z + (1/z))
    return a
fi = np.linspace(0, 2*np.pi)
ro = 1.3
eta = 0.2

plt.plot(z(ro, fi, eta).real, z(ro, fi, eta).imag)
plt.plot(w(z(ro, fi, eta)).real, w(z(ro, fi, eta)).imag)
```
