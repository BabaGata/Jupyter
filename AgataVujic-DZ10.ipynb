### Zadatak 1
Sluƒçajna varijabla ${X}$ slijedi normalnu raspodjelu sa srednjom vrijednosti ${\mu = 50}$ i standardnom
devijacijom ${\sigma = 2}$. Kolika je vjerojatnost da se ${X}$ naƒëe izmeƒëu 47 i 54?

```python
import scipy.stats

print(round(scipy.stats.norm.cdf(x = 54, scale = 2, loc = 50) - 
            scipy.stats.norm.cdf(x = 47, scale = 2, loc = 50),2)*100,'%')
```

### Zadatak 3
Sredi≈°nji graniƒçni teorem ka≈æe da su srednje vrijednosti dovoljno velikih uzoraka neke raspodjele
raspodjeljene prema normalnoj (Gaussovoj) raspodjeli bez obzira na to kakva je originalna raspodjela
iz koje te uzorke uzimamo. U to ƒáemo se uvjeriti na primjeru ${X^2}$ raspodjele.
1. Nacrtajte tu raspodjelu za df=3 stupnja slobode i uvjerite se da je asimetriƒçna oko srednje vrijednosti.
2. Uzmite jedan uzorak od milijun toƒçaka te raspodjele i uvjerite se da je standardna devijacija ${\sigma = \sqrt{2 ùëëùëì}}$.
3. Uzmite 500 uzoraka od po 400 toƒçaka svaki i napravite listu srednjih vrijednosti tih uzoraka. Uvjerite se da je standardna devijacija vrijednosti u listi ${\sigma/\sqrt{400}}$, kako tra≈æi teorem.
4. Nacrtajte histogram srednjih vrijednosti u listi i uvjerite se (superponiranjem krivulje normalne razdiobe) da su one zaista raspodjeljene prema normalnoj razdiobi srednje vrijednosti ${\mu = df = 3}$ i standardne devijacije ${\sigma/\sqrt{400}}$.

```python
from scipy import *
import scipy.stats
import numpy as np
import matplotlib.pyplot as plt

df = 3
ras = scipy.stats.chi2.rvs(df, size=10**6)
print('1) srednja vrijednost =',scipy.stats.chi2.rvs(3, size=10**4).mean())
fig, ax = plt.subplots(figsize=[10,7])
ax.hist(ras, bins=100)

sd = np.sqrt(2*df)
print('2)', round(np.std(ras), 2) == round(sd, 2))

lista = [scipy.stats.chi2.rvs(df, size=400).mean() for i in range(500)]
print('3)',round(np.std(lista), 2) == round(sd/np.sqrt(400), 2))

xs = np.linspace(2.5, 3.5)
fig, ex = plt.subplots(figsize=[10,7])
ex.hist(lista, bins=35)
ex.plot(xs, 12*scipy.stats.norm.pdf(xs, loc=3., scale=sd/np.sqrt(400)))
```
